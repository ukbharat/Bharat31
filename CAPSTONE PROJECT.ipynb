```python
import pandas as pd

# Sample data for hotel room information (two cities, two hotels each)
data = {
    'Sno': [1, 2, 3, 4, 5, 6, 7, 8],
    'Hotel name': ['Hotel A', 'Hotel A', 'Hotel B', 'Hotel B',
                   'Hotel C', 'Hotel C', 'Hotel D', 'Hotel D'],
    'Room Type': ['Standard', 'Deluxe', 'Standard', 'Suite',
                  'Standard', 'Deluxe', 'Standard', 'Suite'],
    'Single Bed': [2, 1, 2, 1, 2, 1, 2, 1],
    'Double Bed': [0, 1, 0, 2, 0, 1, 0, 2],
    'Prices': [100, 150, 120, 250, 110, 160, 130, 270],
    'City': ['Dubai', 'Dubai', 'London', 'London',
             'Paris', 'Paris', 'New York', 'New York']
}

# Extend the sample data for all 12 cities and hotels (you can add data for the remaining cities and hotels)
cities = ['Dubai', 'London', 'Paris', 'New York', 'Tokyo', 'Sydney', 
          'Los Angeles', 'Beijing', 'Mumbai', 'Rome', 'Cairo', 'Toronto']

for city in cities:
    for hotel in ['Hotel A', 'Hotel B', 'Hotel C', 'Hotel D']:
        data['Sno'].append(len(data['Sno']) + 1)
        data['Hotel name'].append(hotel)
        data['Room Type'].append('Standard')
        data['Single Bed'].append(2)
        data['Double Bed'].append(1)
        data['Prices'].append(120)
        data['City'].append(city)

# Create a DataFrame
df = pd.DataFrame(data)

# Export the DataFrame to a CSV file
df.to_csv('Data3.csv', index=False)

# Display the data frame
print(df)
```

```python
import sqlite3
import pandas as pd

# Create a temporary SQLite database
conn = sqlite3.connect(':memory:')

# Sample data for hotels (assuming the table name is 'hotels')
data = [
    (1, 'Hotel A', 'Sydney', 'Location A', 100, 8.5, '4 stars', 'Excellent'),
    (2, 'Hotel B', 'London', 'Location B', 150, 9.0, '5 stars', 'Superb'),
    (3, 'Hotel C', 'Sydney', 'Location C', 80, 7.5, '3 stars', 'Good'),
    (4, 'Hotel D', 'Paris', 'Location D', 120, 8.0, '4 stars', 'Very Good'),
]

# Create a table and import the data
conn.execute('''
CREATE TABLE hotels (
    sno INT,
    "Hotel name" TEXT,
    cities TEXT,
    location TEXT,
    "number of reviews" INT,
    "user ratings" FLOAT,
    "star rating" TEXT,
    "review score" TEXT
)
''')

conn.executemany('INSERT INTO hotels VALUES (?, ?, ?, ?, ?, ?, ?, ?)', data)

# Run the SQL query
query = '''
SELECT *
FROM hotels
WHERE cities = 'Sydney' AND "user ratings" > 5;
'''

result = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Print the result
print(result)
```
